```python
import pandas as pd
```

```python
Data= pd.read_excel(r"C:\Users\educa\Desktop\TableA.xlsx",header=None)
Data
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
df= np.transpose(Data)
```

```python
y=df.iloc[:, 1]
x=df.iloc[:, 0]
```

```python
import pandas as pd
df['z'] = y.apply(lambda x: f'{x:.1%}')
```

```python

plt.figure(figsize=(10,6))

#for i, txt in enumerate(df['z']):
 #   plt.annotate(txt, (x[i], y[i]), textcoords="offset points", xytext=(0,10), ha='center')
    
#plt.plot(x, y, marker='o', linestyle='-', color='gray')
fig = plt.figure(figsize=(9, 5))

ax =  fig.add_subplot()

#ax.plot(x, y, marker='o', linestyle='-', color='gray')
# turn off the right spine/ticks
ax.spines['right'].set_color('none')
ax.yaxis.tick_left()

# set the y-spine
ax.spines['bottom'].set_position('zero')

# turn off the top spine/ticks
ax.spines['top'].set_color('none')
ax.xaxis.tick_bottom()

ax.plot(x, y, marker='o', linestyle='-', color='gray')

for i, txt in enumerate(df['z']):
    plt.annotate(txt, (x[i], y[i]), textcoords="offset points", xytext=(0,10), ha='center')
    
plt.xlabel('Years')  
plt.ylabel('Percentage change from previous years')  
plt.title('GDP GROW RATES AT CONSTANT 2017 PRICES')

#plt.legend()
#plt.grid(True)
plt.savefig('TableA.png')
plt.show()
```

# GDP by kind of Activity #

```python
import pandas as pd
```

```python
Data= pd.read_excel(r"C:\Users\educa\Desktop\TableB.xlsx", index_col=0)
```

```python
Data
```

```python
# Set the figure size - handy for larger output
from matplotlib import pyplot as plt
plt.rcParams["figure.figsize"] = [10, 4]
# Set up with a higher resolution screen (useful on Mac)
%config InlineBackend.figure_format = 'retina'
```

```python
#Data.plot(kind="bar")
Data_new = Data.drop(['GROSS DOMESTIC PRODUCT (GDP)'],axis=1)
Data_new.plot(kind="bar",stacked=True)
Data["GROSS DOMESTIC PRODUCT (GDP)"].plot(kind='line', use_index=False, linestyle='-', marker='o', color='gray', legend='GDP')
plt.title("GDP by Kind of Activity at current prices")
plt.xlabel("Years")
plt.ylabel("Billions Rwf")
plt.savefig('TableB.png')
plt.show()
```

# Urban CPI #

```python
import pandas as pd
```

```python
Data = pd.read_excel(r"C:\Users\educa\Desktop\TableD.xlsx", header=None)
```

```python
Data
```

```python
y=Data.iloc[:, 1]
x=Data.iloc[:, 0]
```

```python
import matplotlib.pyplot as plt
labels = x
sizes = y
pie = plt.pie(sizes,autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.title("Urban CPI Base Year 2014")
plt.xlabel("Years")
plt.legend(bbox_to_anchor=(0.85, 1), loc='upper left', labels=labels)
plt.savefig('TableD.png')
plt.show()
```
